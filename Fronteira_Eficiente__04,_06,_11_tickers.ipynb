```python
from finquant.portfolio import build_portfolio
import matplotlib.pyplot as plt
```

#04 tickers | 0.89 Sharpe Ratio

```python
tickers4 = ['SBSP3.SA','BOVA11.SA','COGN3.SA']
```

```python
pf = build_portfolio(names=tickers4, data_api='yfinance')
```

```python
pf.risk_free_rate = 0.0525
```

```python
pf.properties()
```

```python
pf.mc_optimisation(50000)
```

```python
plt.figure(figsize=(18,8))
pf.mc_plot_results()
pf.ef_plot_efrontier()
pf.ef_plot_optimal_portfolios()
pf.plot_stocks()

plt.title("Fronteira Eficiente", fontsize=22)
plt.xlabel('Volatilidade', fontsize=12)
plt.ylabel('Retorno Esperado', fontsize=12)
```

```python
pf.mc_properties()
```
